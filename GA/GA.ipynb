```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from math import floor
```

```python
def  create_firstPopulation(count, length, equation):#создание первой популяции
    p_matrix = [ [0]*length for i in range(count) ]
    for i in range(count):
        for j in range(length):
            p_matrix[i][j] = random.randint(1, equation[len(equation)-1])
    return p_matrix    
    
    
def get_fitCoef(matrix, equation, count, length):#вычисляем коэф-т выживаемости, те подставляем в уравнеие и смотрим насколько д-ко от значения
    coef = [None] * count
    c = 0
    for i in range(count):
        for j in range(length):
            c += matrix[i][j]*equation[j]
        coef[i] = c      
        c =0
        coef[i] = abs(coef[i]- equation[len(equation)-1])
    return coef


def get_answ(coef, cr):
    for i in range(len(coef)):
        if coef[i]==0:
            cr = True


def get_parent(prob):
    parents_matrix = [ [0]*2 for i in range(prob)]
    for i in range(prob):
        for j in range(2):
            parents_matrix[i][j] = random.randint(0, prob-6)
            while parents_matrix[i][0]==parents_matrix[i][1]:
                parents_matrix[i][1]=random.randint(0, prob-6)
    return parents_matrix

def get_population(parentsPairs_matrix, length, count, parents_matrix):#создание новой популяции
    population_matrix = [None]*count
    for i in range(count):
        population_matrix[i] = parents_matrix[parentsPairs_matrix[i][0]][0:3]+parents_matrix[parentsPairs_matrix[i][1]][3:4]
    return population_matrix

def get_mutation(population_matrix, count, length, equation):#мутируем парамаетры1 или 2, 3 или 4
    for i in range(count):
        population_matrix[i][random.randint(0, 1)] = random.randint(1, equation[len(equation)-1])
        population_matrix[i][random.randint(2, 3)] = random.randint(1, equation[len(equation)-1])
    return population_matrix

def random_equation(equation_length):
    equation = [None]*(equation_length+1)
    for i in range (equation_length):
        equation[i] =  random.randint(1, 5)
    equation[equation_length] = random.randint(20, 30)
    return equation
    
if __name__=="__main__":
    equation = random_equation(4) #создание уравнения
    print("уравнение:")
    print(equation)
    count = 10
    length = 4
    avg = 0
    cr = False
    index = 0
    #создание первой популяции
    matrix = create_firstPopulation(count, length, equation)
    coef = get_fitCoef(matrix, equation, count, length)  
    count = count + 5 #популяция увеличивается с арифметической прогрессией (d = 5)
    parentsPairs_matrix = get_parent(count) 
   
    population = get_population(parentsPairs_matrix, length, count, matrix)
    
    while cr == False:#проверка на наличие решения в популяции
        matrix = population
        coef = get_fitCoef(matrix, equation, count, length)
        for i in range(len(coef)):
            if coef[i]==0:
                cr = True
                index = i
                print(population)
                print ("решение нашлось в популяции:")
                print((len(population)-5)/5)
                print("решение:")
                print(population[index])
                break
        
        count = count +5
        parentsPairs_matrix = get_parent(count)            
        population = get_population(parentsPairs_matrix, length, count, matrix)
        population =  get_mutation(population, count, length, equation)
   
```
